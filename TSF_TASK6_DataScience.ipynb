## Prediction using Decision Tree Algorithm
Create and visualize a Decision Tree Classifier for the 'Iris' dataset. This classifier empowers accurate prediction of the right class when fed with new data.


```python
# Necessary Libraries and imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python

# Ignoring warnings
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading the Iris dataset
iris_df=pd.read_csv('Iris.csv')
```

```python
# Exploring the dataset
iris_df.info()
```

```python
#Exploring first 5 values in the dataset
iris_df.head()
```

```python
# Descriptive Summary
iris_df.describe(include='all')
```

```python
# Unique Species of iris
iris_df['Species'].unique()
```

```python
# In-Depth Analysis
# Firstly Checking if there are any NULL values
iris_df.isnull().sum()
```

```python
#Visualization of Iris data using seaborn library
sns.pairplot(iris_df.drop('Id',axis=1))
```

```python
 # checking the correlation between attributes
plt.figure(figsize=(10,5))
sns.heatmap(iris_df.drop('Id',axis=1).corr(),annot=True,cmap='coolwarm')
```

```python
# Implementing the Decision Tree
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Splitting the dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# DataScaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Training the model on train data
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
# Prediction on test data
y_pred = classifier.predict(X_test)
```

```python
result = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
result
```

```python
# Visualizing of the Tree Model
from sklearn.tree import plot_tree

feature_names=['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
class_names=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
features_names = list(iris_df.columns[:4])

plt.figure(figsize=(40,25))
plot_tree(classifier,feature_names=feature_names,class_names=class_names,filled=True);
```

```python
# Calculating Accuracy
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# Accuracy report
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Output Prediction


```python
# labeling the features
labels = {0:"setosa",1:"versicolor",2:"virginica"}
# Sample Input
new_input = np.array([4.8,7.8,3.9,3.6]).reshape(1,-1)
```

```python
# Predicted Output
predicted_output = classifier.predict(new_input)[0]
```
